```python
# Example 1 - Inner Join
mortality_data = pd.merge(df, adult_mortality, how = 'inner', left_on = ['Unnamed: 0'], right_on = ['Unnamed: 0'])
mortality_data.shape

# print the shapes
print(mortality_data.shape)
```
